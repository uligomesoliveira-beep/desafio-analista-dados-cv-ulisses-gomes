```python
import pandas as pd
import requests
#Objetivo deste código é acessar a API de feriados públicos e carregar algumas linhas prévias do data frame para verificação exploratória.

url_feriados = "https://date.nager.at/api/v3/publicholidays/2024/BR"#Busca somente os feriados brasileiros

resposta_feriados = requests.get(url_feriados)

if resposta_feriados.status_code == 200:
    dados_feriados = resposta_feriados.json()
    df_feriados = pd.DataFrame(dados_feriados)
    
    df_feriados['date'] = pd.to_datetime(df_feriados['date']).dt.date #A base 'public holiday's' tem a coluna 'date' em formato string, não compatível com o formato doda biblioteca 'pandas', então convertemos para 'datetime' para extrairmos a data.
    
    print("Tabela de Feriados Brasileiros - 2024")
    print(df_feriados[['date', 'localName', 'name']].head())
else:
    print(f"Erro ao acessar a API de Feriados. Código: {resposta_feriados.status_code}") #mensagem de erro caso a API não seja acessada com sucesso.
```

**PERGUNTA 4**-Qual foi a temperatura média em cada mês? e **PERGUNTA 5**-Qual foi o tempo predominante em cada mês nesse período?
**Observação**: Diferentemente das demais respostas, aqui optou-se por responder ambas conjuntamente em um mesmo código devido a similaridade da lógica a ser empregada e para melhorar o output do script.

```python
# Objetivo do código: Analisar o clima predominante em cada mês de 2024 utilizando 'df_clima' e o 
# dicionário de códigos climáticos do WMO presente no repositório do GitHub, renderizando os resultados de forma visual.
import pandas as pd
import requests
from IPython.display import display, HTML # Importando a biblioteca para renderizar HTML's presentes no WMO'Code como imagens

# ==============================================================================
# 1. ACESSAR A API DE CLIMA (Open-Meteo)
# ==============================================================================
url_clima = "https://archive-api.open-meteo.com/v1/archive?latitude=-22.9064&longitude=-43.1822&start_date=2024-01-01&end_date=2024-12-01&daily=temperature_2m_mean,weather_code&timezone=America/Sao_Paulo"
resposta_clima = requests.get(url_clima)

if resposta_clima.status_code == 200:
    dados_diarios = resposta_clima.json()['daily']
    df_clima = pd.DataFrame(dados_diarios)
    df_clima['time'] = pd.to_datetime(df_clima['time'])
    df_clima['mes'] = df_clima['time'].dt.month
else:
    print("Erro ao acessar API de Clima")

# ==============================================================================
# 2. ACESSAR A API DO DICIONÁRIO WMO (GitHub Gist)
# ==============================================================================
url_gist = "https://gist.githubusercontent.com/stellasphere/9490c195ed2b53c707087c8c2db4ec0c/raw/" # Importa o repositório do WMO'Code direto do GitHub

# 👇 ESTA É A LINHA QUE FALTAVA PARA FAZER O DOWNLOAD DOS ÍCONES 👇
resposta_gist = requests.get(url_gist) 

if resposta_gist.status_code == 200:
    wmo_github = resposta_gist.json()
    print("✅ Dicionário WMO do GitHub importado com sucesso!\n")
else:
    print("Erro ao carregar o Gist do GitHub")
    wmo_github = {}

# ==============================================================================
# 3. PREPARAÇÃO DOS DADOS E CÁLCULO DAS MÉDIAS MENSAIS
# ==============================================================================
meses = {1: 'Janeiro', 2: 'Fevereiro', 3: 'Março', 4: 'Abril', 5: 'Maio', 6: 'Junho', 
         7: 'Julho', 8: 'Agosto', 9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'}

# Respostas da pergunta 4: Qual foi a temperatura média de cada mês em 2024 no Rio de Janeiro?
# Respostas da pergunta 5: Qual foi o tempo predominante (sol, chuva, nublado, etc.) em cada mês de 2024 no Rio de Janeiro?
print("_"*60)
print("CLIMA E TEMPO PREDOMINANTE NO RIO DE JANEIRO (JAN-DEZ/2024)")
print("_"*60)

# Agrupando por mês, tirando a média da temperatura e a moda (mais frequente) do clima
df_mensal = df_clima.groupby('mes').agg(
    temp_media=('temperature_2m_mean', 'mean'),
    tempo_predominante=('weather_code', lambda x: x.mode()[0])
).reset_index()

# ==============================================================================
# 4. RENDERIZAÇÃO VISUAL EM HTML
# ==============================================================================
html_output = "" # Variável que irá acumular o código HTML gerado para cada mês, que será renderizado no final.

for index, row in df_mensal.iterrows():
    nome_mes = meses.get(row['mes'], f"Mês {int(row['mes'])}")
    codigo_wmo_str = str(int(row['tempo_predominante']))
    
    # Buscando a descrição e a imagem no dicionário que importamos do GitHub
    if codigo_wmo_str in wmo_github:
        desc_tempo = wmo_github[codigo_wmo_str]['day']['description']
        imagem_url = wmo_github[codigo_wmo_str]['day']['image']
    else:
        desc_tempo = "Desconhecido"
        imagem_url = ""

    # Criando bloco visual para cada mês usando HTML e CSS inline
    html_output += f"""
    <div style="display: flex; align-items: center; margin-bottom: 15px; font-family: Arial, sans-serif;">
        <img src="{imagem_url}" alt="{desc_tempo}" style="width: 50px; height: 50px; margin-right: 15px; background-color: #2c3e50; border-radius: 8px; padding: 5px;">
        <div>
            <strong style="font-size: 16px;">{nome_mes}</strong><br>
            🌡️ Temp. Média: {row['temp_media']:.1f}°C<br>
            ☁️ Tempo: {desc_tempo}
        </div>
    </div>
    """

# Exibindo o resultado final bonitão na tela
display(HTML(html_output))
```

**PERGUNTA 6**-Qual foi o tempo e a temperatura média em cada feriado de 01/01/2024 a 01/08/2024?

```python
#Objetivo do código: Cruzar os dados de feriados e clima com 'merge' para analisar o clima em cada feriado e aplicar o "estilo carioca" de aproveitar o dia de acordo com o clima e o tempo
from IPython.display import display, HTML

data_limite = pd.to_datetime('2024-08-01')
df_feriados_filtro = df_feriados[df_feriados['date'] <= data_limite].copy()
# Criamos uma coluna só de data no df_clima para o merge funcionar perfeitamente
df_clima['date_only'] = df_clima['time'].dt.normalize() #Cria coluna 'date_only' com apenas a data (sem hora) para facilitar o merge com os feriados

df_analise = df_feriados_filtro.merge(df_clima, left_on='date', right_on='date_only', how='inner')

# Aplicando as Regras do Carioca (Temp >= 20 E WMO <= 2)
# Nota: WMO 0 (Clear), 1 (Mainly Clear), 2 (Partly Cloudy). WMO 3+ já é nublado ou chuva.
df_analise['aproveitavel'] = (df_analise['temperature_2m_mean'] >= 20) & (df_analise['weather_code'] <= 2) #Notas do código WMO para definir se o dia é "aproveitável" para ir à praia ou não, de acordo com a temperatura e o tipo de clima.
#

html_q6 = "<h3 style='color: #333; font-family: Arial;'>CLIMA EM CADA FERIADO (JAN a AGO/2024)</h3>" 

for index, row in df_analise.iterrows():
    data_f = row['date'].strftime('%d/%m/%Y')
    codigo_wmo_str = str(int(row['weather_code']))
    
    if codigo_wmo_str in wmo_github: #Importa o repositório do WMO'Code direto do GitHub
        desc_tempo = wmo_github[codigo_wmo_str]['day']['description']
        imagem_url = wmo_github[codigo_wmo_str]['day']['image']
    else:
        desc_tempo = "Unknown"
        imagem_url = ""
        
    # Textos e cores baseadas no critério de "aproveitável"
    if row['aproveitavel']:
        aproveitavel_str = "☀️ <b>DEU PRAIA!</b>" #Adicionando e cores e ícones de acordo com o tempo
        cor_fundo = "#e8f5e9"
    else:
        aproveitavel_str = "🌧️ <b>FICOU EM CASA</b>"
        cor_fundo = "#ffebee" 
    
    html_q6 += f"""
    <div style="display: flex; align-items: center; margin-bottom: 10px; font-family: Arial, sans-serif; background-color: {cor_fundo}; padding: 10px; border-radius: 8px; color: #333;">
        <img src="{imagem_url}" alt="{desc_tempo}" style="width: 45px; height: 45px; margin-right: 15px; background-color: #2c3e50; border-radius: 8px; padding: 5px;">
        <div style="flex-grow: 1;">
            <strong style="font-size: 15px;">{data_f} - {row['localName']}</strong><br>
            🌡️ Temp: {row['temperature_2m_mean']:.1f}°C | ☁️ Weather: {desc_tempo}
        </div>
        <div style="font-size: 14px;">
            {aproveitavel_str}
        </div>
    </div>
    """
display(HTML(html_q6))
```

**PERGUNTA 8**-Qual foi o feriado "mais aproveitável" de 2024?

```python
#Objetivo do código: Filtrar os feriados "aproveitáveis" para identificar qual foi o "Mais Aproveitável" de 2024
df_bons = df_analise[df_analise['aproveitavel'] == True].copy()
if len(df_bons) > 0:
    melhor_feriado = df_bons.sort_values(by=['weather_code', 'temperature_2m_mean'], ascending=[True, False]).iloc[0]
    codigo_wmo_str = str(int(melhor_feriado['weather_code']))
    desc_tempo = wmo_github.get(codigo_wmo_str, {}).get('day', {}).get('description', 'Unknown')
    
    print(f"\nResultado: O feriado 'Mais Aproveitável' de 2024 foi:")
    print(f"   🏆 {melhor_feriado['localName']} ({melhor_feriado['date'].strftime('%d/%m/%Y')})") #adiciona ícone para deixar a resposta mais visual
    print(f"   🌡️ Temperatura: {melhor_feriado['temperature_2m_mean']}°C | ☁️ Condição: {desc_tempo}")
```
